## Задача №1

### Часть 1

Построить в общих осях графики для:

- $\alpha = 1, \beta = 1$

- $\alpha = 2, \beta = 1$

- $\alpha = 1, \beta = 2$

```python
#Версия python 3.7.4
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np 

a = [1, 2, 1]
b = [1, 1, 2]
x1 = np.arange(-5, 0, 0.2)
x2 = np.arange(0.2, 5, 0.1)

def f1(x):
    return (x**(b[0]) + a[0]**(b[0])) / (x**(b[0]))

def f2(x):
    return (x**(b[1]) + a[1]**(b[1])) / (x**(b[1]))

def f3(x):
    return (x**(b[2]) + a[2]**(b[2])) / (x**(b[2]))


plt.figure(figsize=(13,8))
plt.plot(x1, f1(x1), 'r', label = '$\\alpha = 1, \\beta = 1$')
plt.plot(x1, f2(x1), 'b', label = '$\\alpha = 2, \\beta = 1$')
plt.plot(x1, f3(x1), 'g', label = '$\\alpha = 1, \\beta = 2$')

plt.plot(x2, f1(x2), 'r')
plt.plot(x2, f2(x2), 'b')
plt.plot(x2, f3(x2), 'g')

plt.legend()

plt.xlabel('x', fontsize = 18)
plt.ylabel('f(x)', fontsize = 18) 
plt.title('График функции $f(x) = \\frac{ x^{\\beta} + \\alpha ^{\\beta}}{x^{\\beta}}$', fontsize = 18) 

plt.savefig('1.svg')
plt.show()
```

### Часть 3

На том же графике сделать 1 врезку, демонстрирующую поведение графиков при удалении x от 0 к $-\infty$.

Необходимо продемонстрировать возможность (или невозможность) пересечений и стремление функций. Так же нанесите на графики прямую f(x) = 0.

Цвет линий на врезках и основном графике должен быть одинаковым для одних и тех же значений $\alpha$ и $\beta$.

```python
#Версия python 3.7.4
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np 
import math

a = [1, 2, 1]
b = [1, 1, 2]

x4 = np.arange(-5, 0, 0.2)
xinf = np.arange(-1000, 0, 10)
xinf1 = np.arange(-10000, -9900, 10)
y = 0
def f1(x):
    return (x**(b[0]) + a[0]**(b[0])) / (x**(b[0]))

def f2(x):
    return (x**(b[1]) + a[1]**(b[1])) / (x**(b[1]))

def f3(x):
    return (x**(b[2]) + a[2]**(b[2])) / (x**(b[2]))

def f4(x):
    return x-x
    
fig = plt.figure(figsize=(13,8))

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x4, f1(x4), 'r', label = '$\\alpha = 1, \\beta = 1$')
axes1.plot(x4, f2(x4), 'b', label = '$\\alpha = 2, \\beta = 1$')
axes1.plot(x4, f3(x4), 'g', label = '$\\alpha = 1, \\beta = 2$')
axes1.plot(x4, f4(x4), 'black', label = 'f(x) = 0')
axes1.legend()
axes1.set_xlabel('x', fontsize = 18)
axes1.set_ylabel('f(x)', fontsize = 18) 
axes1.set_title('График функции $f(x) = \\frac{ x^{\\beta} + \\alpha ^{\\beta}}{x^{\\beta}}$', fontsize = 18) 

axes2 = fig.add_axes([0.2, 0.60, 0.3, 0.3])
axes2.plot(xinf, f1(xinf), 'r', label = '$\\alpha = 1, \\beta = 1$')
axes2.plot(xinf, f2(xinf), 'b', label = '$\\alpha = 2, \\beta = 1$')
axes2.plot(xinf, f3(xinf), 'g', label = '$\\alpha = 1, \\beta = 2$')
axes2.plot(xinf, f4(xinf), 'black', label = 'f(x) = 0')
axes2.legend()
axes2.set_xlabel('x', fontsize = 18)
axes2.set_ylabel('f(x)', fontsize = 18) 
axes2.set_title('График функции f(x) от 0 до -$\\infty $', fontsize = 15) 

axes3 = fig.add_axes([0.6, 0.60, 0.3, 0.3])
axes3.plot(xinf1, f1(xinf1), 'r', label = '$\\alpha = 1, \\beta = 1$')
axes3.plot(xinf1, f2(xinf1), 'b', label = '$\\alpha = 2, \\beta = 1$')
axes3.plot(xinf1, f3(xinf1), 'g', label = '$\\alpha = 1, \\beta = 2$')
axes3.legend()
axes3.set_xlabel('x', fontsize = 18)
axes3.set_ylabel('f(x)', fontsize = 18) 
axes3.set_title('График функции f(x) от 0 до -$\\infty $', fontsize = 15) 

plt.savefig('3.svg')
plt.show()
```

## Задача №2

Построить в общих осях графики для:

- $\alpha = 1, \beta = 0$

- $\alpha = 0, \beta = 0$

- $\alpha = 0, \beta = 1$

- $\alpha = 0, \beta = 2$

- $\alpha = 1, \beta = -1$

- $\alpha = 2, \beta = -1$

Сделайте выводы о поведении графиков

```python
#Версия python 3.7.4
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np 
import math

a = [1, 0, 0, 0, 1, 2]
b = [0, 0, 1, 2, -1, -1]

x1 = np.arange(-20, 0, 0.1)
x2 = np.arange(0.1, 20, 0.1)

#x4 = np.arange(-5, 0, 0.2)
#xinf = np.arange(-1000, 0, 10)
#xinf1 = np.arange(-10000, -9900, 10)
#y = 0

def f1(x):
    return (x**(b[0]) + a[0]**(b[0])) / (x**(b[0]))

def f2(x):
    return (x**(b[1]) + a[1]**(b[1])) / (x**(b[1]))

def f3(x):
    return (x**(b[2]) + a[2]**(b[2])) / (x**(b[2]))

def f4(x):
    return (x**(b[3]) + a[3]**(b[3])) / (x**(b[3]))

def f5(x):
    return (x**(b[4]) + a[4]**(b[4])) / (x**(b[4]))

def f6(x):
    return (x**(b[5]) + a[5]**(b[5])) / (x**(b[5]))
    
fig = plt.figure(figsize=(13,8))

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x1, f1(x1), 'r', label = '$\\alpha = 1, \\beta = 0$')
axes1.plot(x1, f2(x1), 'b', label = '$\\alpha = 0, \\beta = 0$')
axes1.plot(x1, f3(x1), 'g', label = '$\\alpha = 0, \\beta = 1$')
axes1.plot(x1, f4(x1), 'y', label = '$\\alpha = 0, \\beta = 2$')
axes1.plot(x1, f5(x1), 'black', label = '$\\alpha = 1, \\beta = -1$')
axes1.plot(x1, f6(x1), 'm', label = '$\\alpha = 2, \\beta = -1$')


axes1.legend()
axes1.set_xlabel('x', fontsize = 18)
axes1.set_ylabel('f(x)', fontsize = 18) 
axes1.set_title('График функции $f(x) = \\frac{ x^{\\beta} + \\alpha ^{\\beta}}{x^{\\beta}}$', fontsize = 18) 

axes1.plot(x2, f1(x2), 'r')
axes1.plot(x2, f2(x2), 'b')
axes1.plot(x2, f3(x2), 'g')
axes1.plot(x2, f4(x2), 'y')
axes1.plot(x2, f5(x2), 'black')
axes1.plot(x2, f6(x2), 'm')

plt.savefig('4.svg')
plt.show()
```

Графики получившихся функций - прямые.
Графики функций,где $\beta = 0$ накладываются друг на друга(красный и синий),т.к. являются одинаковыми.
Аналогичная ситуация с зеленым и желтым графиками,только из-за $\alpha = 0$. 
При x<0 график устремляется в минус бесконечность.
При x>0 график устремляется в плюс бесконечность.
